# Test pour avoir une liste

```python
cur = collection_po.find().limit(5)
list(cur)
```

# Classement par année

```python
cur = collection_po.find().sort([("annee", -1)])
list(cur)
```

# Classement par sanction type Avertissement

```python
cur = collection_po.find({"sanction": "Avertissement"})
list(cur)
```

# Nombre de type de sanction

```python
cur = collection_po.aggregate([{"$group":{"_id":"$sanction"}},{"$count":"type_sanction"}])
list(cur)
```

# Classement par Genre et le motif 

```python
cur = collection_po.aggregate([{"$group":{"_id":"$sexe", "motif" : {"$avg":"$motif"}}}])
list(cur)
```

# Recherche du nombre de sanctions par motif

```python
cur = collection_po.aggregate([{"$group":{"_id":"$motif", "type_sanction" : {"$avg":"$nombre"}}},{"$sort":{"type_sanction": -1}}])
list(cur)
```

# Nombre d'hommes ayant pour motif "Violence" avec un nombre inférieur à 20

```python
cur = collection_po.find({"$and":[{"sexe": "H"}, {"motif": "Violences"}, {"groupe":{"$lt":'20'}}]}).count()
print(cur)
```
